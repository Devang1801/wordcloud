```python
import os
import numpy as np
import pandas as pd
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1=pd.read_csv('E:/project/dataset/t243 cleaned/final_T243_zero.csv', low_memory=False)
df1.head(10)
```

```python
df2=df1.iloc[:,58]
df2
```

```python
text_data = ' '.join([str(elem) for elem in df2])
text_data
```

```python

wordcloud = WordCloud(max_font_size=100, max_words=10000,stopwords=STOPWORDS, background_color="white", width=1500, height=1500).generate(text_data)

def invert_wc(wc):
    keys = list(wc.words_.keys())
    values = list(wc.words_.values())
    values.reverse()
    wc.words_ = dict(zip(keys, values))
    return wc
invert_wc(wordcloud)


wordcloud = WordCloud(max_font_size=100, max_words=10000,stopwords=STOPWORDS, background_color="white", width=1500, height=1500).generate_from_frequencies(wordcloud.words_)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
